```python
mbpol = pd.read_csv("mbpol_sim.dat", delim_whitespace=True).set_index("Step")
```

```python
qtip = pd.read_csv("qtip4p_sim.dat", delim_whitespace=True).set_index("Step")
```

```python
del mbpol['Progress (%)']
del qtip['Progress (%)']
del mbpol['Time Remaining']
del qtip['Time Remaining']
```

```python
for col in mbpol.columns:
    plt.figure()
    pd.rolling_mean(mbpol[col], 500).plot(label="mbpol", color="r")
    pd.rolling_mean(qtip[col], 500).plot(label="qtip4pf", color="b")
    plt.legend()
    plt.title(col)
```
