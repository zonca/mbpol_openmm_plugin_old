```python
from glob import glob
```

```python
data = {}
for f in glob("qtip4p_sim_*_beads.dat"):
    data[int(f.split("_")[2])] = pd.read_csv(f, delim_whitespace=True).set_index("Step")
```

```python
data[2].columns
```

```python
plt.figure(figsize=(10, 6))
col = "Kinetic Energy (kJ/mole)"
for k,v in data.iteritems():
        pd.rolling_mean(v[col]/k, 500).plot(label=str(k))
plt.legend(loc=0)
```

```python
mean_col = {}
col = "Kinetic Energy (kJ/mole)"
for k,v in data.iteritems():
        mean_col[k] = v[col].mean()/k
```

```python
pd.Series(mean_col).plot(marker="s")
```
